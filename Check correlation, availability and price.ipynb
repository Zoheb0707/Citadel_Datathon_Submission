```python
import pandas as pd
```

```python
calendar = pd.read_csv(filepath_or_buffer = 'calendar.csv')
```

```python
venues = pd.read_csv(filepath_or_buffer = 'venues.csv')
real_estate = pd.read_csv(filepath_or_buffer = 'real_estate.csv')
demographics = pd.read_csv(filepath_or_buffer = 'demographics.csv')
econ_state = pd.read_csv(filepath_or_buffer = 'econ_state.csv')
listings = pd.read_csv(filepath_or_buffer = 'listings.csv',low_memory=False)
```

```python
calendar = calendar[calendar['available'] == 't']
calendar.dropna(axis=1,inplace=True)
calendar['year'], calendar['month'], calendar['day'] = calendar['date'].str.split('-', 2).str
calendar.drop(['date', 'day', 'metro_area'],axis=1,inplace=True)
calendar['available'] = 1
calendar['price'] = (calendar['price'].str)[1:]
```

```python
list_of_price = []
for i in range(0,len(calendar)):
    price = calendar['price'].iloc[i]
    if ',' in price:
        price = price.replace(",", "")
    list_of_price.append(float(price))
calendar['price'] = list_of_price
```

```python
calendar['price'] = list_of_price
```

```python
grouped_clendar = calendar.groupby(by = ['listing_id', 'year', 'month'])['price', 'available'].sum()
```

```python
grouped_clendar['price'] = grouped_clendar['price']/grouped_clendar['available']
```
